Part-2: Lab task is to visualize the datasets or data using Matplotlib builtin function graphs

```python
people=['shivani', 'gayathri', 'varshitha', 'pavithra', 'chandana']
age=[20,19,18,17,21]
weight=[45,40,30,50,55]
height=[10,11,12,14,15]
import matplotlib.pyplot as plt
plt.scatter(weight, height)
plt.title("relationship betweeen weight and height of patients")
plt.ylabel("height in meters")
plt.xlabel("weight in kgs")
plt.show()
```

Barplot

```python
plt.figure(figsize=(12,4))
plt.title("people's weight in kgs",fontsize=16,fontstyle='italic')
plt.bar(x=people,height=weight,width=0.6,color='orange',edgecolor='k',alpha=0.6)
plt.xlabel("people",fontsize=15)
plt.xticks(fontsize=14,rotation=30)
plt.yticks(fontsize=14)
plt.ylabel("weight (in kgs)",fontsize=15)
plt.show()
```

Histogram

```python
import numpy as np
plt.figure(figsize=(7,5))
plt.hist(weight,color='red',edgecolor='k',alpha=0.75,bins=5)
plt.title("Histogram of patient weight",fontsize=18)
plt.xlabel("weight in kgs",fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```
